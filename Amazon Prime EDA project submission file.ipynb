## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd

# Define the file paths (Update these paths based on your system)
titles_path = r"C:\Users\DELL\Downloads\extracted_files\titles.csv"
credits_path = r"C:\Users\DELL\Downloads\extracted_files\credits.csv"

# Load datasets
df_titles = pd.read_csv(titles_path)
df_credits = pd.read_csv(credits_path)

print("‚úÖ Datasets loaded successfully!")
````

````python
# Fill missing categorical values
df_titles = df_titles.assign(
    description=df_titles["description"].fillna("Unknown"),
    age_certification=df_titles["age_certification"].fillna("Not Rated")
)

# Fill missing numerical values
df_titles = df_titles.assign(
    seasons=df_titles["seasons"].fillna(df_titles["seasons"].median()),
    imdb_score=df_titles["imdb_score"].fillna(df_titles["imdb_score"].mean()),
    imdb_votes=df_titles["imdb_votes"].fillna(df_titles["imdb_votes"].median()),
    tmdb_popularity=df_titles["tmdb_popularity"].fillna(df_titles["tmdb_popularity"].median()),
    tmdb_score=df_titles["tmdb_score"].fillna(df_titles["tmdb_score"].mean())
)

print("‚úÖ Missing values handled successfully without warnings!")
````

````python
columns_to_drop = ["cast", "crew"]  # Dropping cast and crew
df_credits.drop(columns=[col for col in columns_to_drop if col in df_credits.columns], inplace=True, errors="ignore")

print("‚úÖ Unnecessary columns dropped!")
````

````python
df_titles["release_year"] = df_titles["release_year"].astype(int)
print("\n‚úÖ Data Types After Conversion:\n", df_titles.dtypes)
````

````python
df_titles.drop_duplicates(inplace=True)
df_credits.drop_duplicates(inplace=True)
print("‚úÖ Duplicate rows removed!")
````

````python
df = df_titles.merge(df_credits, on="id", how="left")
print("‚úÖ Merged Dataset:")
print(df.head())
````
